## Then, you can group the data by its City column and this new list of County:

```python
by_pe_ci = data2.groupby(['City',permit])
```

```python
agg_counts = by_pe_ci.size().unstack().fillna(0)
agg_counts
```

```python
new_agg_counts = agg_counts[agg_counts['permit']>=200]
```

```python
new_agg_counts['%_of_permits'] = (new_agg_counts['permit']/(new_agg_counts['No permit']+new_agg_counts['permit']))*100
```
